```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample time series data
data = {
    'date': pd.date_range(start='2022-01-01', periods=100, freq='D'),
    'value': [i + (i * 0.1) for i in range(100)]  # Simple increasing trend
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the time series
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['value'], label='Value', color='b')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Rolling mean
df['rolling_mean'] = df['value'].rolling(window=10).mean()

# Plotting the time series with rolling mean
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['value'], label='Original Data', color='b', alpha=0.5)
plt.plot(df['date'], df['rolling_mean'], label='Rolling Mean (10 days)', color='r', linestyle='--')
plt.title('Time Series with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

```

```python
import seaborn as sns

# Seaborn plot with time series data
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['date'], y=df['value'], label='Value', color='b')
sns.lineplot(x=df['date'], y=df['rolling_mean'], label='Rolling Mean', color='r', linestyle='--')

plt.title('Time Series with Seaborn')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

# Create interactive time series plot
fig = px.line(df, x='date', y='value', title='Interactive Time Series')
fig.add_scatter(x=df['date'], y=df['rolling_mean'], mode='lines', name='Rolling Mean', line=dict(dash='dash'))

# Show the plot
fig.show()
```

```python
import statsmodels.api as sm

# Decompose time series into trend, seasonal, and residual components
decomposition = sm.tsa.seasonal_decompose(df['value'], model='additive', period=30)

# Plot the decomposition
decomposition.plot()
plt.show()
```
